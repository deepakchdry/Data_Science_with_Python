```python
import pandas as pd
demo = {'Region':['kanto','johto','hoenn','sinnoh','unova'],
        'Temperature':[23,56,78,45,65],
        'Humidity':[34,54,21,43,11],
        'Rainfall':[21,33,77,44,65]
        }
df = pd.DataFrame(demo)
df
w1, w2, w3 = 0.3, 0.2, 0.5
weight = [w1,w2,w3]

kanto = [23	,34,	21]
johto	= [56,	54	,33]
hoenn = [78,	21,	77]
sinnoh = [45,	43,	44]
unova = [65	,11,	65]
```

```python
%%time
def crop_yield(region,weight):
  result = 0
  for x,y in zip(region,weight):
    result += (x*y)
  return result

crop_yield(kanto,weight)
```

```python
crop_yield(unova,weight)
```

```python
crop_yield(sinnoh,weight)
```

```python
import numpy as np
mat = np.array([[23, 65, 87],
              [91, 80, 64],
              [87, 14, 57],
              [12, 43, 37],
              [62, 90, 70]])

mat1 = np.array([2,3,4])

x = np.dot(mat,mat1)
y = np.matmul(mat,mat1)
z = mat*mat1
w = mat@mat1

print(x,"\n",y,"\n",w,"\n",z)
```
